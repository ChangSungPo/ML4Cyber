## Variables

Variables can be considered containers. You can put anything inside a container, **without specifying the size or type**, which will be needed in Java or C. When **assigning values**, we put the variable to be assigned to on the left hand side (LHS), while the value to plug in on the RHS. LHS and RHS are connected by an equal sign (=), meaning assignment.

````python
# =============================================================================
# VARIABLES IN PYTHON
# =============================================================================
# Unlike Java/C, Python uses DYNAMIC TYPING - no need to declare types!
# The interpreter figures out the type automatically based on the value.
# =============================================================================

# Integer - whole numbers (no decimal point)
x = 3

# Floating point number - numbers with decimals
y = 2.5

# String - text data, enclosed in quotes (single ' or double " both work)
z = "Hello"

# IMPORTANT: Variable names are CASE-SENSITIVE!
# 'z' and 'Z' are completely different variables
Z = "Wonderful!"

# NAMING RULES for variables:
# ✓ Can contain: letters, numbers, underscores
# ✓ Must START with: letter or underscore
# ✗ CANNOT start with a number: 3_x = "456" would be invalid!

# print() function displays the value of variables
print(x)
print(y)
print(z)
print(Z)

# Variables can be REASSIGNED to different types (dynamic typing)
# Here we change Z from a string to a boolean
Z = True

print("New value of Z:", Z)
````

You can do operations on numeric values as well as strings.

````python
# =============================================================================
# ARITHMETIC OPERATIONS
# =============================================================================
# Python supports: +, -, *, /, // (floor div), % (modulo), ** (power)
# =============================================================================

# Adding an integer by a floating point number
# Python automatically converts to float (this is called "type coercion")
# x = 3 (int), y = 2.5 (float) → result is float
w = x + y
print(w)  # Output: 5.5 (notice it's a float!)

# =============================================================================
# STRING CONCATENATION
# =============================================================================
# The + operator joins strings together (concatenation)
# =============================================================================
v = "World!"
hello_world = z + " " + v  # Joining 3 strings: "Hello" + " " + "World!"
print(hello_world)  # Output: Hello World!
````

However, we have a more convenient way of concatenating strings, called **formatting**. Formatting can be thought of a process of filling texts into a pre-defined string template. A formatting string template begins with letter "f" and the variables to be formatted are put in curly brackets in the string.

````python
# =============================================================================
# F-STRINGS (Formatted String Literals) - Python 3.6+
# =============================================================================
# Syntax: f"text {variable} more text"
# - Start with 'f' before the opening quote
# - Put variables inside curly braces {}
# - Much cleaner than concatenation with +
# =============================================================================

# Embedding variables directly in strings
print(f"The concatenated string of '{z}' and '{v}' is '{hello_world}'!")

# FORMAT SPECIFIERS for numbers:
# {value:.2f} → float with 2 decimal places
# {value:d}   → integer
# {value:,}   → number with comma separators
print(f"The sum of x and y is {w:.2f}")  # Output: 5.50
````

You can think of strings as a sequence of characters (or a list of characters, see the next section). In this case, indices and bracket notations can be used to access specific ranges of characters.

````python
# =============================================================================
# STRING INDEXING AND SLICING (Very Important!)
# =============================================================================
# Visual guide for the string "abcdefg":
#
#  Index:    0   1   2   3   4   5   6
#           [a] [b] [c] [d] [e] [f] [g]
# Negative: -7  -6  -5  -4  -3  -2  -1
#
# SLICING SYNTAX: string[start:end]  ← end is EXCLUSIVE (not included)!
# =============================================================================

letters = 'abcdefg'

# Get the second character (index 1, since Python starts at 0)
second_letter = letters[1]
print(second_letter)  # Output: b

# Negative indices count from the END (-1 = last element)
last_letter = letters[-1]
print(last_letter)  # Output: g

# SLICING: Get characters from index 1 to 3 (end index 4 is excluded!)
# letters[1:4] → indices 1, 2, 3 → "bcd"
letters_substr_1 = letters[1:4]
print(letters_substr_1)  # Output: bcd

# Omitting start/end:
# letters[:3] → from beginning to index 2 (first 3 characters)
# letters[3:] → from index 3 to end
letters_substr_2 = letters[3:]
print(letters_substr_2)  # Output: defg
````

## Built-in Data Types
In this section, we will go over common built-in data types in Python. Most of the built-in data types are primitive types, meaning that they cannot be further decomposed into simpler ones.

### Numbers

````python
# =============================================================================
# NUMERIC TYPES IN PYTHON
# =============================================================================
# int   - Integers (whole numbers): 1, -5, 1000
# float - Floating point (decimals): 3.14, -0.5, 3. (note: 3. = 3.0)
# complex - Complex numbers: 2j+5 means 5 + 2i in math notation
# =============================================================================

# Integer type
x = 1
# type() function returns the data type of any value/variable
print(type(x))  # Output: <class 'int'>

# Floating point number type
# Note: 3. is shorthand for 3.0
y = 3.
print(type(y))  # Output: <class 'float'>

# Complex number type (used in engineering/physics)
# In Python, 'j' represents the imaginary unit (i in math)
z = 2j+5  # This is 5 + 2i
print(type(z))  # Output: <class 'complex'>
````

### List
A list is a sequential and mutable data structure. You can store anything in a list, even if contained elements are of different types.

````python
# =============================================================================
# LISTS - The Most Commonly Used Data Structure
# =============================================================================
# Lists are ORDERED, MUTABLE (can be changed), and allow DUPLICATE elements
# Syntax: [element1, element2, element3, ...]
# =============================================================================

# Create a list with square brackets []
a_list = [1, 2, 3, 4]

# len() returns the number of elements
print(f"Length of a_list is: {len(a_list)}")  # Output: 4

# Indexing works just like strings (0-based!)
# Remember: 3rd element = index 2
print(f"3rd element of a_list is: {a_list[2]}")  # Output: 3

# Slicing: [start:end] - end is EXCLUSIVE
# [1:-1] means from index 1 to second-to-last (excluding last)
print(f"Slice a sub-list from a_list: {a_list[1:-1]}")  # Output: [2, 3]

# Built-in sum() adds all numeric elements
print(f"The sum of a_list is: {sum(a_list)}")  # Output: 10
````

### Set
Set is a mutable data structure that resembles a collection in math. It is non-sequential and all of its elements are unique. It is useful for de-duplicating values.

````python
# Initialize an empty set
a_set = set()
print(a_set)

# Initialize a set from an existing list or tuple
a_set = set(a_tuple)
print(a_set)

# Add an element to the set
a_set.add(4)
print("After adding element 4, a_set is now:")
print(a_set)

# Remove an element from the set
a_set.remove(2)
print("After removing element 2, a_set is now:")
print(a_set)

# Check if a value is in the set or not
a_set_has_1 = 1 in a_set
print(f"a_set has element 1: {a_set_has_1}")
````

### Dictionary
Dictionary is a mutable data structure that contains key-value pairs. A key in the dictionary must correspond to only one value.

````python
# Initialize a dictionary with curly brackets
empty_dict = {}
print(empty_dict)

a_dict = {1: "a", "key": "value", b"bool_prop": True}
print(a_dict)

# Add or update a key-value pair
a_dict[2] = "b"
a_dict[1] = "aa"
print("After updating, a_dict is now:")
print(a_dict)

# Remove a key-value pair from dictionary
del a_dict[b"bool_prop"]
print("After removing key b'bool_prop', a_dict is now:")
print(a_dict)
````

````python
# Get keys of the dictionary with `keys` method
# Note `keys` returns an iterator, so we collect all the keys into a list
a_dict_keys = list(a_dict.keys())
print(f"Keys of a_dict: {a_dict_keys}")

# Get values of the dictionary with `values` method
a_dict_values = list(a_dict.values())
print(f"Values of a_dict: {a_dict_values}")

# `items` method returns an iterator for visiting all key-value pairs
for key, value in a_dict.items():
    print(f"Value of a_dict[{key}] is {value}")
````

## Control statements

### If-Else

````python
# =============================================================================
# IF-ELIF-ELSE STATEMENTS
# =============================================================================
# Syntax:
#   if condition1:
#       # code runs if condition1 is True
#   elif condition2:
#       # code runs if condition1 is False AND condition2 is True
#   else:
#       # code runs if ALL conditions above are False
#
# IMPORTANT: Don't forget the colon : after each condition!
# IMPORTANT: Code blocks are defined by INDENTATION (usually 4 spaces)
# =============================================================================

sum_ = 6.0

if sum_ == 0:
    print("sum_ is 0")
elif sum_ < 0:
    print("sum_ is less than 0")
else:
    # str() converts a number to string for concatenation
    print("sum_ is above 0 and its value is " + str(sum_))
````

### For Loop
For loop can be used for visiting elements in an iterable object, such as string, bytes, list, tuple and set.

````python
# =============================================================================
# FOR LOOPS
# =============================================================================
# Syntax: for variable in iterable:
#             # code to execute for each element
#
# An "iterable" is anything you can loop over:
# strings, lists, tuples, sets, dictionaries, range(), etc.
# =============================================================================

# Looping through a string (each character becomes 'letter')
iter_string = "abcde"
print("Iterating through a string:")
for letter in iter_string:
    print(letter)

print("")

# Looping through a list (each element becomes 'item')
# Note: lists can contain mixed types!
iter_list = [1, 2.2, False, None, b"hi!", ("hello", "world")]
print("Iterating through a list:")
for item in iter_list:
    print(item)
````

Range is a virtual number sequence type that helps us to write C-style for loops:

````python
# Construct range with end, begin index and step
# (Latter two are optional)
range_with_end = range(4)
range_with_begin_end = range(2, 5)
range_with_begin_end_step = range(1, 10, 2)

print(f"range(4) is equivalent to list {list(range_with_end)}")
print(f"range(2, 5) is equivalent to list {list(range_with_begin_end)}")
print(f"range(1, 10, 2) is equivalent to list {list(range_with_begin_end_step)}")
````

### List Comprehension
List comprehension provides a more concise way to transform a data structure.

````python
# Create a new list by adding 1 to each item
old_list = [2, 3, 5]
new_list = [x+1 for x in old_list]
print(f"Old list: {old_list}, new list: {new_list}")

# Filter elements in the list
num_list = [2, 3, 5, 8, 9]
even_list = [x for x in num_list if x%2==0]
print(f"Old list: {num_list}, filtered list with even numbers only: {even_list}")
````

### While Loop
While loop executes its body until its condition no longer holds. It is useful when you don't know the exact number of iterations (e.g. keep running learning algorithm before reaching a threshold of accuracy). You can break out of `for` and `while` loop with **break** statement, or jump to the next iteration with **continue** statement:

````python
a = 1
b = 1

# Fibonacci sequence loop
while True:
    print(f"a = {a}")
    
    # Update next item of the sequence
    a, b = b, a
    a += b

    # Do not print `b` if `a` is smaller than 8
    if a <= 8:
        continue
    print(f"b = {b}")
    
    # Break out of the loop if `b` is larger than 20
    if b > 20:
        break
````

### Define our own functions
Note that we are not limited to built-in functions and we can make our own functions. Before that, however, we need to be clear on the syntax of a function:

```python
def func_name(arg1, arg2, ...):
    """ Description of the function here. """
    #####################
    # Do something here #
    #####################
    return output
```

* Be careful of the indentation!
* Description of function is recommended, but not required.
* Returning output is not required, in which case the function will return `None`.

````python
def my_sum(list_to_sum):
    """ Compute the sum of an iterable by calling built-in sum function. """
    return sum(list_to_sum)

my_sum(range(5))
````

````python
def my_sum_using_loop(list_to_sum):
    """ Compute the sum of iterable with for loop. """
    sum_ = list_to_sum[0]
    for item in list_to_sum[1:]:
        sum_ += item
    return sum_

my_sum_using_loop(range(5))
````

#### Broadcasting / Repeating / Reshaping
The term **broadcasting** refers to the ability of NumPy to match elements of arrays with different shapes using specific rules during arithmetic operations. Code utilizing broadcasting are much faster than code that uses for loop to do computations.



````python
a = np.array([[0, 0, 0], [10, 10, 10], [20, 20, 20], [30, 30, 30]])
b = np.array([0, 1, 2])

# `np.empty_like` creates an uninitialized array with the shape of another array
result1 = np.empty_like(a)

# Addition through for loop
for i in range(4):
    result1[i, :] = a[i, :] + b
# Same addition using broadcasting
result2 = a + b[None, :]

print(result1, "\n")
print(result2)
````

Now let's practice **broadcasting** through an exercise. We will compute the Kronecker product of two vectors, and broadcasting is done as shown below:



````python
# Two vectors of different lengths
x = np.array([1, 2, 3])      # Shape: (3,)
y = np.array([1, 2, 3, 4])   # Shape: (4,)

## [ TODO ] ============================================================
# Compute the OUTER PRODUCT (Kronecker product) of x and y using broadcasting.
#
# The result should be a 4x3 matrix where result[i,j] = y[i] * x[j]
#
# HINTS:
# 1. Broadcasting needs compatible shapes. Current shapes don't match!
# 2. Use None (or np.newaxis) to add a dimension:
#    - x[None, :] reshapes x from (3,) to (1, 3) - a row vector
#    - y[:, None] reshapes y from (4,) to (4, 1) - a column vector
# 3. When you multiply (1,3) * (4,1), NumPy broadcasts to (4,3)
#
# STEP BY STEP:
#   x = [1, 2, 3]     →  x[None, :] = [[1, 2, 3]]        shape (1,3)
#   y = [1, 2, 3, 4]  →  y[:, None] = [[1], [2], [3], [4]]  shape (4,1)
#   
#   Multiply them: result[i,j] = y[i] * x[j]
#
# EXPECTED OUTPUT:
# [[ 1  2  3]
#  [ 2  4  6]
#  [ 3  6  9]
#  [ 4  8 12]]
# ======================================================================
z = x*y[:, None]  # Replace with: x[None, :] * y[:, None] (or similar)
print(z, "\n")
````

### Pandas
**Pandas** is a Python library that provides advanced data structures and tools for data-driven analysis. We will introduce basic usage of `DataFrame` below, which is very similar to a table in the database.

````python
# =============================================================================
# CREATING A SAMPLE DATAFRAME
# =============================================================================
# We'll generate random employee data to practice Pandas operations
# =============================================================================

from random import Random
import pandas as pd

# Random number generator with fixed seed for reproducibility
# (31415926 = digits of pi - a common choice for reproducible results)
rand = Random(31415926)

# Sample data pools
names = ["Albert", "John", "Richard", "Henry", "William"]
surnames = ["Goodman", "Black", "White", "Green", "Joneson"]
salaries = [500*rand.randint(10, 30) for _ in range(10)]

def generate_random_person(names, surnames, salaries):
    """
    Generate a random person dictionary with name, surname, and salary.
    
    Args:
        names: List of possible first names
        surnames: List of possible last names
        salaries: List of possible salary values
    
    Returns:
        Dictionary with keys: "name", "surname", "salary"
    """
    return {
        "name": rand.sample(names, 1)[0],
        "surname": rand.sample(surnames, 1)[0],
        "salary": rand.sample(salaries, 1)[0]
    }

def generate_people(k):
    """Generate a list of k random person dictionaries."""
    return [generate_random_person(names, surnames, salaries) for _ in range(k)]

# Create DataFrame with 50 random employee profiles
df = pd.DataFrame(generate_people(50), columns=["name", "surname", "salary"])

# head(n) shows first n rows - great for previewing data!
df.head(10)
````

### Matplotlib
Matplotlib is a Python library for plotting and visualizing data.

````python
import matplotlib.pyplot as plt

# Create curve data
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot two series (two different curves)
plt.plot(x, y_sin, label="Sine")
plt.plot(x, y_cos, label="Cosine")

# Set axis labels
plt.xlabel("x")
plt.ylabel("y")

# Set plot title
plt.title('Sine and Cosine')

# Show legend
plt.legend()

# Complete and show the plot
plt.show()
````
